# Harmonic Oscillator
## Part (a)
We have to plot the potential function given by
$$\frac{1}{2}m\omega^2x^2$$
where $m=\omega=\hbar=1$.

```python
def V(x):
    '''
    The potential function for an harmonic oscillator. Returns the value of V at x.
    '''
    
    return 0.5 * x ** 2

# Plotting the potential of harmonic oscillator

# Range of x axis
x_max = 3
N = 1000 # Number of points to plot the potential
x = np.linspace(-x_max, x_max, N)

# The Plot
plt.plot(x, V(x))
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('The harmonic oscillator potential')

# Show the plot
plt.show()
```

## Part (b)
Here since we need hermite polynomial and the nth derivative would make it inaccurate, I thought of using the recurance relationship to find out the hermite polynomial though it would be time consuming which also could be the case of taking derivatives n times. Also, taking the derivatives n times would reduce the number of points by 1.

```python
def H(x, n):
    '''
    Rucursively implemented hermite polynomial.
    '''
    if n == 0:
        return 1
    elif n == 1:
        return 2 * x
    else:
        return 2 * x * H(x, n-1) - 2 * (n-1) * H(x, n-2)
    
def psi_unnormalized(x, n):
    '''
    Returns the value of the unnormalized eigenfunction of the harmonic oscillator of the nth state. 
    '''
    
    return np.exp(- x**2 / 2.) * H (x, n)
```

We know that the wavefunction in the forbidden region decays exponentially, so to integrate 
$$ \int^\infty_{-\infty} \left | \psi_n(x)  \right |^2 = 1 $$
we can choose the limits to be slightly greater than the value at $E=V(x)$ as $E = n + 0.5$ in our case.

```python
# Limit finder
def lim_find(n, buffer=3):
    '''
    Returns what should be the limit for the integral.
    '''
    
    Energy = n + 0.5
    return np.sqrt(Energy) + buffer

def integrate(n, a, N=10000):
    '''
    Returns the norm of the unnormalised wavefunction. Since psi ** 2 is symmetric based on the 
    potential plot we can integrate from 0 to upper limit.
    Uses the Simpson method to integrate the function. So odd number of points have been considered.
    N here represents the number of intervals.
    '''
    h = a/N
    W = np.ones(N+1)*h/3
    W[1::2] *= 4
    W[2:-1:2] *= 2
    W[0] = W[-1] = h/3

    return 2 * np.sum(W * (psi_unnormalized(np.linspace(0,a,N+1), n) ** 2), axis=0)
```

```python
# We find the normalization constant
def normalize(n):
    '''
    Returns the normalization constant for nth state.
    '''
    a = lim_find(n)
    A = integrate(n, a)

    return np.sqrt(1. / A)

# For n = 2 
A2 = normalize(2)
# For n = 5 
A5 = normalize(5)


print('For n=2, normalization constant :', A2)
print('For n=5, normalization constant :', A5)
```

## Part (c)

```python
## Using the nonormalization constant we plot the normalized states.
N = 1000 # Number of points for discreatization
x = np.linspace(-6, 6, N)

def psi_2(x):
    '''
    Normalized 2nd state for harmonic oscillator
    '''
    
    return A2 * psi_unnormalized(x, 2)

def psi_5(x):
    '''
    Normalized 2nd state for harmonic oscillator
    '''
    
    return A5 * psi_unnormalized(x, 5)

_ = plt.figure(figsize=(10,3)) # For nice graph shape
plt.plot(x, psi_2(x), 'r-', label='2nd Stationary State')
plt.plot(x, psi_5(x), 'b-', label='5th Stationary State')

plt.title('The stationary states')
plt.xlabel('x')
plt.ylabel('$\psi_n(x)$')

plt.legend()
plt.show()
```

## Part (d)
I will be using bisection method as newton rapson may not converge and may give other solution rather than the close one. Also, we have to find the derivative of psi (which could be done numerically with errors) and so I have used bisection.

```python
def bisection(start, end, func, MaxSteps=50, acc=1e-15):
    '''
    Finds the root for the function 'func' in the given start end interval and returns it.
    Will return None if it can't find the root.
    '''
    z = acc + 1
    steps = 0
        
    # If no root exists, then None is returned.
    if func(start) * func(end) > 0:
        return None
    
    while z > acc:
        mid = (start + end) / 2.
        if func(start) * func(mid) > 0:
            start = mid
        elif func(end) * func(mid) > 0:
            end = mid
        else:
            return mid

        z = np.fabs((start - end) / (start + end))
        steps += 1
        
    return mid
```

Based on plotting the stationary states we have 2 nodes for 2nd stationary states and 5 for 5th one. We can figure the intervals from the graph above.

```python
# For the 2nd stationary states
roots_2 = np.zeros((2,), dtype=float)

roots_2[0] = bisection(-2, 0, psi_2)
roots_2[1] = bisection(0, 2, psi_2)

# For the 2nd stationary states
roots_5 = np.zeros((5,), dtype=float)

roots_5[0] = bisection(-3, -1.9, psi_5)
roots_5[1] = bisection(-1.6, -0.5, psi_5)
roots_5[2] = bisection(-0.5, 0.5, psi_5)
roots_5[3] = bisection(0.5, 1.5, psi_5)
roots_5[4] = bisection(2, 2.5, psi_5)
```

```python
# Display the nodes
print('Nodes for the 2nd stationary states.')
for i in range(roots_2.size):
    print(f'Node {i}: {roots_2[i]}')

print('')
print('Nodes for the 5th stationary states.')
for i in range(roots_5.size):
    print(f'Node {i}: {roots_5[i]}')
```

We see that the nodes are symmetric and also close to what it looks likes its accurate.

## Part (e)
We only have to find the kinetic energy of the state of $n=2$. Now we can use the central double differenciation formula to find out the derivative of $\psi_2(x)$.

```python
def der2(x, func, h=1e-6):
    '''
    Finds the 2nd derivative of the function 'func' at the point x.
    '''
    
    return (func(x + h) + func(x - h) - 2. * func(x)) / (h ** 2)

def kinetic_energy(x, h=1e-6):
    '''
    Returns the kinetic enrgy of the psi_2 state.
    '''

    return - 0.5 * der2(x, psi_2, h) / psi_2(x)
```

```python
# Plotting the kinetic energy
N = 10000 # Number of points
start = -0.5
end = 0.5
x = np.linspace(start, end, N)

# The Plot
plt.plot(x, kinetic_energy(x))
plt.xlabel('x')
plt.ylabel('KE')
plt.title('The Kinetic energy of 2nd stationary state.')

# Show the plot
plt.show()
```

## Part (f)
We solve the the 2nd order differential equation by splitting it into 2 1D equations

```python
# The RK4 method
def rk4(x, dx, t, tf, h=1e-3):
    '''
    Performs the 4th order Runge-Kutta method, for an independant variable t and a range
    of dependant varaibles x_i (x).
    
    x  : The variables (x_0 = x, x_1 = y ...) 
    dx : Differential equations system dx/dt = .... where x is for all x's above
    t  : Intial time when to start solving system
    tf : Stops solving when t >= tf.
    h  : The step size (Default = 1 x 10^(-6)) (did this so only have to change in one place)
    '''
    
    # Storing the initial condition as it would be required too.
    t_list = [t]
    x_list = [x]
    
    while(t < tf):
        # Finding the coefficients k1,....
        k1 = h * dx(t, x)
        k2 = h * dx(t + h/2., x + k1/2.)
        k3 = h * dx(t + h/2., x + k2/2.)
        k4 = h * dx(t + h, x + k3)

        # Update step
        t = t + h
        x = x + k1/6. + k2/3. + k3/3. + k4/6.

        # Save the data to plot later
        # Lists are faster at appending than numpy arrays
        t_list.append(t)
        x_list.append(x)
        
    return np.array(t_list), np.array(x_list)
```

```python
# The 2 differntial equation with E = 2.5
def f1(x, y, E=2.5):
    return np.array([y[1], (x**2)*y[0] - 2*E*y[0]])

# Initial conditions
x = np.array([0.0055, 0.0193])
t = -4
tf = 4

t_list, x_list = rk4(x, f1, t, tf)
plt.plot(t_list, x_list[:,0], 'b-', label='Using SE Equation')
plt.plot(t_list, psi_2(t_list), 'r--', label='Using the definition')
plt.xlabel('x')
plt.ylabel('$\psi_2(x)$')

plt.legend()
plt.show()
```

We see that the graphs overlap each other.

# Numerical Derivatives and Intergration 
## Part (A)
We have to compare the numerical and analytical derivative of the 
$$ 1 + \frac{1}{2} \tanh(2x)$$ 
and its derivative is
$$  \frac{1}{2} \mathrm{sech}^2(2x) $$

```python
def func(x):
    '''
    The function to be differentiated
    '''
    
    return 1. + 0.5 * np.tanh(2 * x)

def central_der(x, func, h=1e-10):
    '''
    Calculates the central difference derivative of function 'func'
    '''
    
    return (func(x+h) - func(x-h)) / (2. * h)

def anal_der(x):
    '''
    Analytical derivative of the given function.
    '''
    
    return 1. / (np.cosh(2. * x) ** 2)
```

```python
## Using the nonormalization constant we plot the normalized states.
N = 1000 # Number of points for discreatization
x = np.linspace(-2, 2, N)

plt.plot(x, anal_der(x), 'r-', label='Analytical')
plt.plot(x, central_der(x, func), 'b--', label='Numerical')

plt.title('Derivatives of $f(x)$')
plt.xlabel('x')
plt.ylabel('$f(x)$')

plt.legend()
plt.show()
```

## Part (B)
### Part (a)
We have to use the trapezoidal rule to find
$$ E(x) = \int_0^x e^{-t^2} dt $$

```python
def f(x):
    '''
    The function inside to be integrated
    '''
    
    return np.exp(- x*x)

def E(x, N=10000):
    '''
    Integrates 'func' using trapezoidal rule. N is the number of points.
    '''
    t = np.linspace(0, x, N)
    
    h = x / (N-1)
    W= [h] * N
    W[0] = W[-1] = h / 2.

    return np.sum(W * f(t))
```

```python
# The valuse of x we need E(x)
start = 0
end = 3
step = 0.1

x = np.arange(start, end, step)

# Integrating to find E(x)
Soln = [ E(y) for y in x ]

print('E(x) values are :', Soln)
```

```python
# The Plot
plt.plot(x, Soln)
plt.xlabel('x')
plt.ylabel('E(x)')
plt.title('E(x) vs x')

# Show the plot
plt.show()
```

## Part (C)
### Part (a)
Given $n=1$ and $a=1.5$ so
$$ F(x) = \int_0^\pi \cos(x^{1.5}\cos(t)) \sin^{3}(t) dt $$


```python
def toBeIntegrated(x, t):
    '''
    Returns the function to be integrated.
    '''
    return np.cos((x ** 1.5) * np.cos(t)) * np.sin(t) ** 3

def F(x, N=10000):
    '''
    Returns F(x) using the trapezoidal rule as I have used more points it is small and compact.
    '''
    h = np.pi/(N-1)
    W = np.ones(N)*h
    W[0] = W[-1] = 0.5*h

    return np.sum(W*toBeIntegrated(x,np.linspace(0,np.pi,N))) / np.pi
```

```python
## Using the nonormalization constant we plot the normalized states.
N = 100 # Number of points for discreatization
x = np.linspace(0, 5, N)
Soln = [ F(y) for y in x ]

# The Plot
plt.plot(x, Soln)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('F(x) vs x')

# Show the plot
plt.show()
```

# Solving Poisson's equation for 2 sources
Here we have to solve the Poisson's equation for 2 charges. I have used here the Jocobi method which can also have relaxation to speed it up. Though the Gauss-Seidel does it in approximately half the cycles, the double for loop and the nature of the algorithm (preventing vectorization as we have to use the previous value) slows it down really significantly. Relaxation can help it perform better but with vectorization and relaxation Jacobi is an appropriate choice in Python unless we use external libraries to boost.

Coming to the question we have to solve the Poissons equation. We take the surrounding points to get the derivative and subtract the non homegenous part.

$$
U_{i,j}^{k+1} = \frac{1}{4} \left( U^k_{i+1,j} + U^k_{i-1,j} + U^k_{i,j+1} + U^k_{i,j-1} \right) - \frac{1}{4} dx  dy \left(-\frac{\rho_{i,j}}{\epsilon_0}\right)
$$

```python
def relaxation(Grid, NonHomogenous=None, dx=1, dy=1, MinDiff=1e-5, overRelax=0.0, 
               MaxStep=1e6, verbose=False):
    '''
    Uses the relaxation method to solve the Laplace equation untill difference is 
    less than MinDiff or it doesn't converge until MaxStep is reached.
    
    It is the same as Jacobi Method.
    Over relaxation gives overflow as it diverges at some point and is numerically unstable.
    I could be wrong but pls check it out.
    '''
    
    diff = MinDiff + 1 # for the first while loop iteration
    steps = 0
    
    if NonHomogenous is None:
        NonHomogenous = np.zeros(Grid.shape, dtype=np.float64)
    
    while steps < MaxStep and diff > MinDiff:
        Copy_g = Grid.copy()
        Grid[1:-1, 1:-1] = (1.0 + overRelax) / 4.0 * (Copy_g[:-2, 1:-1] + Copy_g[2:, 1:-1] 
                            +  Copy_g[1:-1, :-2] + Copy_g[1:-1, 2:]) - (dx * dy / 4.
                            * NonHomogenous[1:-1, 1:-1])
        Grid[1:-1, 1:-1] -= overRelax * Copy_g[1:-1, 1:-1]
        
        diff = np.max(np.abs(Grid-Copy_g))
        
        steps += 1
        
        if verbose:
            print(steps, diff) # If you want to see progress
        
    
    if steps >= MaxStep:
        print("\nMax steps reached, could not converege\n")
```

```python
def boundary(Grid):
    '''
    Gives the boundary condition.
    All the walls are grounded except the first one is maintained at 1.0 V
    '''
    Grid[:,-1] = 0.0
    Grid[:,0] = 0.0
    Grid[0,:] = 0.0
    Grid[-1,:] = 1.0 # Top wall
```

```python
# Discreatiation : I used number of intervals as 100 other than points because
# then the coordinates would make sense as we are including 0 to 100 cm.
length = 100 # cm
Nx = 100
Ny = 100
dx = length / Nx
dy = length / Ny

Grid = np.zeros((Nx+1, Ny+1), dtype=np.float64)
X, Y = np.meshgrid(np.linspace(0, 100, len(Grid)), np.linspace(0, 100, len(Grid)))
boundary(Grid)
HeatMap = plt.contourf(X, Y, Grid, levels=500)
cbar = plt.colorbar(HeatMap)
cbar.set_label('Voltage (V)')
plt.title('The initial condition')
plt.xlabel('Length (cm)')
plt.ylabel('Length (cm)')
plt.show()
```

```python
# Setting up the Non Homogenous part for the charges.
def charges(Grid, dx=1, dy=1, epsilon=1):
    '''
    Sets up the charge densities for the non homogenous part of Laplace's Equation.
    Epsilon_naught is 1.
    '''
    NonHomogenous = np.zeros(Grid.shape, dtype=np.float64)
    NonHomogenous[int(20//dx),int(20//dy)] = 1 / (dx * dy) / epsilon  # q1 = 1 at (20 cm, 20 cm)
    NonHomogenous[int(80//dx),int(80//dy)] = -1 / (dx * dy) / epsilon # q2 = -1 at (80 cm, 80 cm)
    
    return -1 * NonHomogenous

NonHomogenous = charges(Grid, dx, dy)
```

```python
# Solving the system, overrelaxation diverges quickly after a certain point so its ignored
relaxation(Grid, NonHomogenous, dx, dy)

HeatMap = plt.contourf(X, Y, Grid, levels=500)
cbar = plt.colorbar(HeatMap)
cbar.set_label('Voltage (V)')
plt.title('The solution')
plt.xlabel('Length (cm)')
plt.ylabel('Length (cm)')
plt.show()
```

# Differentiation - spectral methods
We have to use specral method to differentiate and then compare it the finite difference method. The difference between the both is that the spectral method looks at the function globally to differentiate unlike the finite difference which take ths surrounding points.

The given function is 
$$
f(x) = \begin{cases}
1 - |x| & \text{if } x \in [-1,1] \\
0 & \text{otherwise}
\end{cases}
$$

```python
def func(x):
    '''
    Returns the functional value at x for the given function in the question.
    '''
    try:
        n = x.size
    except:
        x = np.array(x)
        
    return np.squeeze(np.max(np.vstack([1. - np.abs(x), np.zeros(x.size)]), axis=0))
```

```python
# We will be use the inbuilt function for fft and inverse fft as those were not taught in class
# It is possible to use integral and do fourier transform but its slow and previous assignments
# were based on it.
from scipy.fftpack import fft, ifft, fftfreq

# Discreatization of the domain
N = 1000
lim = 2.
x = np.linspace(-lim, lim, N)

# i times the wave vectors
ik = 1j * np.hstack([np.arange(0, N//2+1), np.arange(-N//2+1,0)]);

u = func(x)
u_hat = fft(u)
v_hat = ik * u_hat * np.pi / lim
v = np.real(ifft(v_hat)) 
```

```python
plt.plot(x, v)
plt.title('Derivative using spectral methods')
plt.show()
```

```python
# Finite difference method
def central_diff(x, func, h=1e-5):
    '''
    Calculates the numerical derivative using finite difference method.
    '''
    return (func(x+h) - func(x-h)) / (2. * h)

y = central_diff(x, func)
```

```python
# Comparing them
plt.plot(x, func(x), label='Original Function')
plt.plot(x, v, label='Spectral Method')
plt.plot(x, y, label='Finite Difference')

plt.title('Comparision')
plt.legend()
plt.show()
```

We see that in this case the central difference is giving a better result near the edges but with the spectral method using FFT we observe the Gibbs phenomenon. In general a spectral method method gives a better answer for less rapidly changing functions.
